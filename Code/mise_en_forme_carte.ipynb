## 2 - Rajout de colonnes utiles pour la carte

```python
carte_donations_farmers['total_amount_per_state'] = (
    carte_donations_farmers['other_total_amount'] +
    carte_donations_farmers['rep_total_amount'] +
    carte_donations_farmers['dem_total_amount']
)

carte_donations_farmers['total_people_per_state'] = (
    carte_donations_farmers['dem_people'] +
    carte_donations_farmers['rep_people'] +
    carte_donations_farmers['other_people']
)

#Nous utilisons cette donnée là pour avoir le pourcentage par parti dans chaque état
carte_donations_farmers['total_num_per_state'] = (
    carte_donations_farmers['dem_num_donations'] +
    carte_donations_farmers['rep_num_donations'] +
    carte_donations_farmers['other_num_donations']
)
```

```python
#Ici aussi, nous rajoutons des colonnes pour avoir la répartition par parti dans chaque état en pourcentage.
carte_donations_farmers['dem_percentage'] = (carte_donations_farmers['dem_num_donations'] / carte_donations_farmers['total_num_per_state'] * 100).round(1)
carte_donations_farmers['rep_percentage'] = (carte_donations_farmers['rep_num_donations'] / carte_donations_farmers['total_num_per_state'] * 100).round(1)
carte_donations_farmers['other_percentage'] = (carte_donations_farmers['other_num_donations'] / carte_donations_farmers['total_num_per_state'] * 100).round(1)
```
