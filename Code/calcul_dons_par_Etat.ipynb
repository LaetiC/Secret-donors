# 2) Déterminer le parti

```python
# Il y a plusieurs façons d'obtenir le parti à partir du numéro identifiant du comité :
# Soit le comité a été directement relié à un parti dans le fichier des comités
# Soit le comité est relié à un candidat qui lui-même est relié à un parti
# Soit le parti est déductible à partir de mots-clés
# Soit le parti doit être trouvé en faisant une recherche puis en le remplissant à la main
```

```python
# Stocker la liste des comités auxquels ont donné les agriculteurs dans un dataframe
committee_farmers = df_farmers[['CMTE_ID']].drop_duplicates()
```

## b) parti déduit à partir du lien comité/candidat

## c) parti déduit à partir de mots-clés

```python
# Identifier les partis restants en fonction des mots-clés
# Fonction pour assigner le parti en fonction des mots-clés dans CMTE_NM

key_words_Rep =  ["DONALD", "TRUMP", "REPUBLICAN", "REPUBLICANS", "RED", "WINRED"]
key_words_Dem = ["KAMALA", "HARRIS", "DEMOCRAT", "DEMOCRATS", "BLUE", "ACTBLUE"]

def assign_party_based_on_keywords(row):
    if pd.isna(row['PTY']) and pd.notna(row['CMTE_NM']):
        cmte_nm_upper = row['CMTE_NM'].upper()
        if any(keyword in cmte_nm_upper for keyword in key_words_Rep):
            return 'REP'
        elif any(keyword in cmte_nm_upper for keyword in key_words_Dem):
            return 'DEM'
    return row['PTY']

# Appliquer la fonction aux lignes pour compléter les valeurs de PTY restantes
committee_farmers['PTY'] = committee_farmers.apply(assign_party_based_on_keywords, axis=1)
```

## d) parti ajouté à la main

```python
Republicans = ['RIGHT FOR AMERICA', "NATIONAL CATTLEMEN'S BEEF ASSOCIATION POLITICAL ACTION COMMITTEE (NCBA-PAC)", "CALIFORNIA DAIRIES, INC. FEDERAL POLITICAL ACTION COMMITTEE",
               "MAKE AMERICA GREAT AGAIN INC.", "NRCC", "UNITED STATES PEANUT POLITICAL ACTION COMMITTEE (US PEANUT PAC)",
               "PLAINS COTTON GROWERS INC. PAC", "SOUTHEAST COTTON COMMITTEE (SECC) SOUTHERN COTTONGROWERS INC/SE COTTON GINNERS ASSN",
               "TAKE BACK YOUR LIBERTmerY", "NATIONAL SORGHUM PRODUCERS POLITICAL ACTION COMMITTEE",
               "AMERICAN VALUES 2024", "FARMERS' RICE COOPERATIVE FUND", "NATIONAL PECAN FEDERATION INC. POLITICAL ACTION COMMITTEE (PECAN PAC)",
               "PFLUGER VICTORY COMMITTEE", "MAKE LIBERTY WIN", "ALABAMA PEANUT PRODUCERS ASSOCIATION, PEANUTPAC OF ALABAMA"]

Democrats = ["DEMOCRACY SUMMER 2024", "NEWSOM FOR CALIFORNIA GOVERNOR 2022", "SMP"]

committee_farmers.loc[committee_farmers['CMTE_NM'].isin(Republicans), 'PTY'] = "REP"
committee_farmers.loc[committee_farmers['CMTE_NM'].isin(Democrats), 'PTY'] = 'DEM'
```

# 3) Exploration des données

```python
df_farmers = pd.merge(df_farmers, committee_farmers[['CMTE_ID','PTY']], on='CMTE_ID', how='left')
```

```python
df_farmers.head(10)
```

```python
# on calcule pour chaque parti et chaque Etat le nombre de donations, le montant total donné et le nombre de donateurs
donations_summary = df_farmers.groupby(['STATE', 'PTY']).agg(
    num_donations=('TRANSACTION_AMT', 'size'),  # Compter le nombre de donations
    total_amount=('TRANSACTION_AMT', 'sum'),
    people = ('NAME', 'nunique')     # Calculer le nombre unique de donateurs
).reset_index()
donations_summary
```

```python
# Toutes les donations affiliées à un parti autre que républicain sont regroupées dans la catgéorie "OTHER"
donations_summary['PTY'] = donations_summary['PTY'].apply(
    lambda x: x if x in ['DEM', 'REP'] else 'OTHER'
)
```

```python
# on réarrange le dataframe pour qu'il soit plus facilement utilisable sur flourish
reshaped_df = donations_summary.pivot_table(
    index='STATE',  # Rows will represent each state
    columns='PTY',  # Columns will be based on PTY
    values=['num_donations', 'total_amount', 'people'],  # Values to organize
    aggfunc='sum'  # Aggregate if there are duplicates
)

# Flatten the MultiIndex columns for better readability
reshaped_df.columns = ['_'.join(col).lower() for col in reshaped_df.columns]

# Reset the index to make it a clean DataFrame
reshaped_df = reshaped_df.reset_index()

# Rename the columns for clarity (optional)
reshaped_df = reshaped_df.rename(columns={
    'num_donations_dem': 'dem_num_donations',
    'num_donations_rep': 'rep_num_donations',
    'num_donations_other': 'other_num_donations',
    'total_amount_dem': 'dem_total_amount',
    'total_amount_rep': 'rep_total_amount',
    'total_amount_other': 'other_total_amount',
    'people_dem': 'dem_people',
    'people_rep': 'rep_people',
    'people_other': 'other_people',
})

# Display the final reshaped DataFrame
print(reshaped_df)
```

```python
#Transformer floats en integers
reshaped_df = reshaped_df.fillna(0)

for col in reshaped_df.columns:
    if reshaped_df[col].dtype in ['float64', 'int64']:
        reshaped_df[col] = reshaped_df[col].astype(int)

print(reshaped_df)
```
